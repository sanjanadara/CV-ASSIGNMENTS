```python
import sys
import numpy as np
import cv2 as cv

    
window_name = ('Sobel Demo - Simple Edge Detector')
scale = 1
delta = 0
ddepth = cv.CV_16S
    
    

    # Load the image
src = cv.imread("C:\\Users\Prasad\Downloads\CV Assigment-2\wall2.jpeg", cv.IMREAD_COLOR)
    # Check if image is loaded fine
if src is None:
    print ('Error opening image: ')
    
    
    
    
gray = cv.cvtColor(src, cv.COLOR_BGR2GRAY)
(thresh, binary) = cv.threshold(gray, 127, 255, cv.THRESH_BINARY)
grad_x = cv.Sobel(binary, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
grad_y = cv.Sobel(binary, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
abs_grad_x = cv.convertScaleAbs(grad_x)
abs_grad_y = cv.convertScaleAbs(grad_y)
grad = cv.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
    
src1 = cv.GaussianBlur(gray, (7, 7), 0)
(thresh, binary1) = cv.threshold(src1, 127, 255, cv.THRESH_BINARY)
grad1_x = cv.Sobel(binary1, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
grad1_y = cv.Sobel(binary1, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
abs_grad1_x = cv.convertScaleAbs(grad1_x)
abs_grad1_y = cv.convertScaleAbs(grad1_y)
grad1 = cv.addWeighted(abs_grad1_x, 0.5, abs_grad1_y, 0.5, 0)
    
    
src2 = cv.GaussianBlur(gray, (9, 9), 0)
(thresh, binary2) = cv.threshold(src2, 127, 255, cv.THRESH_BINARY)
grad2_x = cv.Sobel(binary2, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
grad2_y = cv.Sobel(binary2, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
abs_grad2_x = cv.convertScaleAbs(grad2_x)
abs_grad2_y = cv.convertScaleAbs(grad2_y)
grad2 = cv.addWeighted(abs_grad2_x, 0.5, abs_grad2_y, 0.5, 0)
    
    
src3 = cv.GaussianBlur(gray, (11, 11), 0)
(thresh, binary3) = cv.threshold(src3, 127, 255, cv.THRESH_BINARY)
grad3_x = cv.Sobel(binary3, ddepth, 1, 0, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
grad3_y = cv.Sobel(binary3, ddepth, 0, 1, ksize=3, scale=scale, delta=delta, borderType=cv.BORDER_DEFAULT)
abs_grad3_x = cv.convertScaleAbs(grad3_x)
abs_grad3_y = cv.convertScaleAbs(grad3_y)
grad3 = cv.addWeighted(abs_grad3_x, 0.5, abs_grad3_y, 0.5, 0)
    
    
src_ =cv.resize(src,(250,370),interpolation=cv.INTER_NEAREST)
grad_ =cv.resize(grad,(250,370),interpolation=cv.INTER_NEAREST)
grad_1=cv.resize(grad1,(250,370),interpolation=cv.INTER_NEAREST)
grad_2=cv.resize(grad2,(250,370),interpolation=cv.INTER_NEAREST)
grad_3=cv.resize(grad3,(250,370),interpolation=cv.INTER_NEAREST)
    
grad_ =cv.cvtColor(grad_,cv.COLOR_GRAY2BGR)
grad_1=cv.cvtColor(grad_1,cv.COLOR_GRAY2BGR)
grad_2=cv.cvtColor(grad_2,cv.COLOR_GRAY2BGR)
grad_3=cv.cvtColor(grad_3,cv.COLOR_GRAY2BGR)

final=np.concatenate((src_,grad_,),axis=1)
final1=np.concatenate((final,grad_1,),axis=1)
final2=np.concatenate((final1,grad_2),axis=1)
final3=np.concatenate((final2,grad_3),axis=1)

cv.imshow('EDGE DETECTION',final3)
cv.waitKey(0)
cv.destroyAllWindows()

number_of_white_pix = np.sum(grad == 255)
number_of_black_pix = np.sum(grad == 0)

number_of_white_pix1 = np.sum(grad1 == 255)
number_of_black_pix1 = np.sum(grad1 == 0)

number_of_white_pix2 = np.sum(grad2 == 255)
number_of_black_pix2 = np.sum(grad2 == 0)

number_of_white_pix3 = np.sum(grad3 == 255)
number_of_black_pix3 = np.sum(grad3 == 0)
  
print('Number of white pixels:', number_of_white_pix)
print('Number of white pixels1:', number_of_white_pix1)
print('Number of white pixels2:', number_of_white_pix2)
print('Number of white pixels3:', number_of_white_pix3)
```
